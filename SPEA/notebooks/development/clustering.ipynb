### Modules imports

```python
points = np.random.rand(10_000 * 2).reshape(10_000, 2)
```

```python
plt.figure(figsize=[16, 16])
sns.set()

plt.scatter(points[:, 0], points[:, 1], c="b")

plt.grid()
```

```python
cluster = KMeans(n_clusters=10)
```

```python
cluster.fit(points)
```

```python
plt.figure(figsize=[16, 16])
sns.set()

plt.scatter(points[:, 0], points[:, 1], c="b")
plt.scatter(cluster.cluster_centers_[:, 0], cluster.cluster_centers_[: ,1], c="r", s=120)

plt.grid()
```
